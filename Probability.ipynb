2. Write a function to estimate the probability of getting at least one "6" in 10 rolls of a fair die.  
  Steps  
      a. Simulate rolling a die 10 times using a loop.  
      b. Track trials where at least one "6" occurs.  
      c. Calculate the proportion of successful trials.


```python
import random
#Tossing a coin 10000 times
def coin_toss(trials=10000):
    heads=0
    tails=0

    for _ in range(trials):
        toss=random.choice(['H','T'])
        if toss =='H':
            heads+=1
        else:
            tails+=1

    print(f"Coin toss: ({trials} trials):")
    print(f"Heads: {heads} ({heads/trials:4f})")
    print(f"Tails: {tails} ({tails/trials:4f})\n")

#Rolling two dice and computing the probability of getting a sum of 7.
def dice_sum(trials=10000):
    count_seven=0

    for _ in range(trials):
        die1=random.randint(1,6)
        die2=random.randint(1,6)
        if die1 + die2==7:
            count_seven+=1

    probability=count_seven/trials
    print(f"Two Dice Sum=7 ({trials} trials:)")
    print(f"Times Sum=7: {count_seven}")
    print(f"Probability: {probability:4f}\n")
```

Conditional Probability and Bayes' Theorem


3. A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn randomly, its color noted, and it is put back into the bag. If this process is repeated 1000 times, write a Python program to estimate:  
  a. The probability of drawing a red ball given that the previous ball was blue.  
  b. Verify Bayes' theorem with the simulation results.  


Steps  
    a. Use random sampling to simulate the process.  
    b. Compute conditional probabilities directly from the data.  


```python
import random
from collections import Counter

def simulate_balls_draw(trials=1000):
    bag=['Red']*5+['Green']*7+['Blue']*8

    draws=[random.choice(bag)for _ in range(trials)]
    count_prev_blue=0
    count_red_given_prev_blue=0

    count_red=0
    count_blue=0
    count_blue_given_red=0
     
    for i in range(1, trials):
        prev=draws[i -1]
        curr=draws[i]

        if prev =='Blue':
            count_prev_blue+=1
            if curr=='Red':
                count_red_given_prev_blue+=1

        if curr=='red':
            count_red+=1
            if draws[i -1]=='Blue':
                count_blue_given_red+=1

        if curr=='Blue':
            count_blue+=1

    prob_red_given_blue = count_red_given_prev_blue / count_prev_blue if count_prev_blue else 0

    p_red = count_red / (trials - 1)
    p_blue = count_blue / (trials - 1)
    p_blue_given_red = count_blue_given_red / count_red if count_red else 0

    bayes_result = (p_blue_given_red * p_red) / p_blue if p_blue else 0

    print("Simulation Results:")
    print(f"  P(Red | Previous was Blue): {prob_red_given_blue:.4f}")
    print("\nBayes' Theorem Verification:")
    print(f"  P(Red): {p_red:.4f}")
    print(f"  P(Blue): {p_blue:.4f}")
    print(f"  P(Blue | Red): {p_blue_given_red:.4f}")
    print(f"  Bayes' estimate of P(Red | Blue): {bayes_result:.4f}")
```

```python
if __name__=='__main__':
    simulate_balls_draw()
```

Continuous Random Variables


Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
  a. A histogram.  
  b. A probability density function (PDF) overlay.  
  Steps  
      a. Use numpy.random.exponential().  
      b. Use matplotlib to create visualizations.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import expon

np.random.seed(42)

mean = 5
scale = mean  # For exponential distribution, scale = 1/lambda = mean
n_samples = 2000

samples = np.random.exponential(scale=scale, size=n_samples)

plt.figure(figsize=(10, 6))

sns.histplot(samples, bins=50, kde=False, stat="density", color="skyblue", label="Histogram")

x = np.linspace(0, np.max(samples), 1000)
pdf = expon.pdf(x, scale=scale)
plt.plot(x, pdf, 'r-', lw=2, label="PDF")

plt.title("Exponential Distribution (mean = 5)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

Central Limit Theorem


 Simulate the Central Limit Theorem by following these steps  
  a. Generate 10,000 random numbers from a uniform distribution.  
  b. Draw 1000 samples of size n = 30.  
  c. Calculate and visualize the distribution of sample means.  
  Steps  
      a. Use numpy.random.uniform().  
      b. Plot both the uniform distribution and the sample mean distribution for comparison.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

population = np.random.uniform(low=0, high=1, size=10000)
sample_means=[]
n_samples = 1000
sample_size = 30

for _ in range(n_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, color='skyblue', stat="density", kde=False)
plt.title("Original Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")

plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=50, color='salmon', stat="density", kde=True)
plt.title("Distribution of Sample Means (n=30)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```
